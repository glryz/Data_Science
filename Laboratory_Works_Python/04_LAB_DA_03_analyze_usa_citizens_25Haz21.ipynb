```python
def replacer(df,value,subtitude):
    for i in df.columns:
        if df[i].isin([value]).any():
            print(i, "Column has changed")
            unknow = df[i] == value
            df.loc[unknow, i] = subtitude
replacer(df, "?", "Unknown")    
```

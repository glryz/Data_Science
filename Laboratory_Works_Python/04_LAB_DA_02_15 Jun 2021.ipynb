**MultiIndex**

```python
outside = ['A', 'A', 'A', 'B', 'B', 'B','C', 'C', 'C']
inside = [1, 2, 3, 2, 3, 4, 3, 4, 5]
multi_index = list(zip(outside, inside))
multi_index
```

```python
hier_index=pd.MultiIndex.from_tuples(multi_index)
```

```python
df2=pd.DataFrame(np.random.randn(9,4) , index=hier_index, columns=['X','Y','Z','H'])
df2
```

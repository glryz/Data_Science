## 1: Plot number of people by gender

```python
df.head(5)
```

```python
gender_count = df['gender'].value_counts()

gender_count
```

```python
#with pandas

gender_count.plot(kind = "bar", figsize = (8,5))
plt.title("Gender Distribution", fontsize = 16)

plt.text

for index, value in enumerate(gender_count):
    plt.text(index, value, str(value), ha = "center")

plt.show()
```

```python
#with matplotlib

fig, ax = plt.subplots(figsize = (8,5))
ax.bar(gender_count.index, gender_count.values)
ax.set_title("Gender Distribution", fontsize = 16)

for index, value in enumerate(gender_count):
    ax.text(index, value, str(value), ha = "center")

plt.tight_layout
plt.show()
```

```python
"""
.most_common(n)
List the n most common elements and their counts from the most
common to the least.  If n is None, then list all element counts
"""

most_common_15 = names_surnames_count.most_common(15)

most_common_15
```

```python
df_most_common = pd.DataFrame(most_common_15, columns = ["name_or_surname_most_frequency", "frequency"])

df_most_common
```

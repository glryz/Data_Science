```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax = ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
    
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax = ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()/tips.day.count()), (p.get_x()+0.3, p.get_height()+1));
    
# get_heigth, barların boyları. bunu day sayısına bölersem..
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax = ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate(round(p.get_height()/tips.day.count(),2), (p.get_x()+0.3, p.get_height()+1));
    
# round(p.get_height()/tips.day.count(),2 ile virgülden sonraki basamak sayısını ayarladım.
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax = ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
    ax_tip.annotate("%"+str(round((p.get_height()/tips.day.count())*100,2)), (p.get_x()+0.1, p.get_height()-5));
    
    # "%"+str ile str metodunu kullanarak başına yüzde koydum.
    
    # (round((p.get_height()/tips.day.count())*100,2) ile yüzde haline getirdim.
    
    # p.get_x()+0.1 ile rakamları biraz daha ortaya çektim.  
```

```python
# plt ile biraz figure u açalım
plt.figure(figsize=(12,6))

y = flights.groupby(["year", "month"]).sum().values
x = flights.index
sns.lineplot(x=flights.index, y=flights.passengers.values)
plt.show
```

```python
# flights verisini pivotla şu şekle getirelim.

flights_wide = flights.pivot("year", "month", "passengers")
flights_wide.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=flights_wide.T)

# yaz aylarında bir yoğunluk var. bunun kuzey yarımkürede olduğunu düşünebilirim.
# yıldan yıla da yaz aylarındaki artış hiç değişmemiş.
```

```python
# C8329 Joseph hocanın kodu

plt.figure(figsize = (12,6))
sns.lineplot(data = flights, x = 'month', y = 'passengers', hue='year')
plt.legend(loc=(1.1,.0))
plt.show()
```

```python
sns.lineplot(data = flights, x = 'month', y = 'passengers')
plt.legend(loc=(1.1,.0))

# bundan yıla göre hue'yu kaldırıp sadece aylara göre plotumuzu görelim, genel bir fikir versin.
```

```python
# palette özelliğini kullandım:

plt.figure(figsize=(12,6))
sns.lineplot(data = flights, x = 'month', y = 'passengers', hue='year', palette="viridis")
plt.legend(loc=(1.1,.0))
```

```python
# pointplot ile yapalım. 
# lineplot'un yaptığı işin aynısını yaptı.

plt.figure(figsize=(12,6))
sns.pointplot(data = flights, x = 'month', y = 'passengers', hue='year', palette="viridis")
plt.legend(loc=(1.1,.0))
```

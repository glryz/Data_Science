# Exercise 9 Aşağıdaki koşulları kullanarak iki veri çerçevesini birleştirin
Aşağıdaki iki sözlüğü kullanarak iki veri çerçevesi oluşturun.

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
```

```python
GC=pd.DataFrame(GermanCars)
JC=pd.DataFrame(japaneseCars)
```

```python
GC
```

```python
JC
```

```python
pd.concat([GC, JC], axis=0, ignore_index = False)
```

```python
pd.concat([GC, JC], axis=0, ignore_index = True)

# ignore_index = True farkı.
```

```python
pd.concat([GC, JC], keys=['GermanCars', 'japaneseCars'], axis=0, ignore_index = False)

# keys belirleyince onları grup ismi olarak index'e koydu. ignore_index = True demediğimiz için de indexler olduğu gibi kaldı
```

```python
pd.concat([GC, JC], keys=['GermanCars', 'japaneseCars'], axis=0, ignore_index = True)

# ignore index yapınca (keysler dahil) önceki belirlenen indexleri gözardı etti ve indexleri tamamen kendi otomatik olarak verdi.
```

```python
GC =pd.DataFrame(GermanCars)
JC =pd.DataFrame(japaneseCars)
```

```python
GC
```

```python
JC
```

```python
pd.concat([GC,JC], keys=["Germany","Japan"])
```

# Exercise 10: Aşağıdaki koşulu kullanarak iki veri çerçevesini birleştirin

Aşağıdaki iki Dikt'i kullanarak iki veri çerçevesi oluşturun, İki veri çerçevesini birleştirin ve ikinci veri çerçevesini ilk veri çerçevesine yeni bir sütun olarak ekleyin.

```python
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

```python
CP =pd.DataFrame(Car_Price)
CH =pd.DataFrame(car_Horsepower)
```

```python
CP
```

```python
CH
```

```python
pd.merge(CP,CH, on="Company")
```

```python
Car_Price2 = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi', 'Tofaş', 'AlfaRomeo'], 'Price': [23845, 17995, 135925 , 71400, 10000, 25000]}
Car_Horsepower2 = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

```python
CP2 = pd.DataFrame(Car_Price2)
HP2 = pd.DataFrame(Car_Horsepower2)
```

```python
CP2
```

```python
HP2
```

```python
pd.merge(CP2,HP2,on="Company", how='outer')
```
